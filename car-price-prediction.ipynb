# Model Building

```python
#define x and y 
X = car.drop(columns= 'Price')
y = car['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 32)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
# ohe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_ ),['name','company','fuel_type']),     #this line does whatever the data will come first it get transform though one hot encoder
                                       remainder = 'passthrough')
                                                               
```

```python
lr = LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)   
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
## lets train the model on differnt random state

scores = []
for i in range(1200):
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state =i)
    lr = LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train, y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state =np.argmax(scores))
lr = LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train, y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# dump this pipeline
import pickle
pickle.dump(pipe,open('LinearRegressionModel.pkl','wb'))
```
